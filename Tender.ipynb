```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time 
import ipywidgets as widgets
from ipywidgets import interact, interactive, fixed, interact_manual,Layout
from tqdm.notebook import tnrange,tqdm
```

```python
Title=[]
Procurementstage = []
Noticestatus=[]
Closing=[]
Contractlocation=[]
Contractvalue=[]
Publicationdate=[]
def get_item_num():
        
        HEADERS = ({'User-Agent':
		'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.51 Safari/537.36',
						'Accept-Language': 'en-US, en;q=0.5'})
        html='https://www.contractsfinder.service.gov.uk/Search/Results'
        # Extract tables
        df = pd.read_html(html)

        r = requests.get(html, headers=HEADERS)
        soup=BeautifulSoup(r.content, "lxml")
        for list_all in soup.find_all('div',class_='gadget-body'):
          for nn in list_all.find_all('div',attrs={'class':'search-result'}):
           title=nn.find('h2').string.strip().replace(',', '')
           procurementstage=nn.find_all('div',attrs={'class':'search-result-entry'})[0].get_text().split()[-1]
           noticestatus=nn.find_all('div',attrs={'class':'search-result-entry'})[1].get_text().split()[-1]
           closing =nn.find_all('div',attrs={'class':'search-result-entry'})[-4].get_text().split()[-1]
           contractlocation =nn.find_all('div',attrs={'class':'search-result-entry'})[-3].get_text().split()[-1]
           contractvalue=nn.find_all('div',attrs={'class':'search-result-entry'})[-2].get_text().split()[-1]
           publicationdate=nn.find_all('div',attrs={'class':'search-result-entry'})[-1].get_text().split()[-1]
 
           Title.append(title if title else '')
           Procurementstage.append(procurementstage if procurementstage else '')
           Noticestatus.append(noticestatus if noticestatus else '')
           Closing.append(closing if closing else '')
           Contractlocation.append(contractlocation if contractlocation else '')
           Contractvalue.append(contractvalue if contractvalue else '')
           Publicationdate.append(publicationdate if publicationdate else '')
      
        return Title
    
num_items=(get_item_num())
#create a dataframe
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
df = pd.DataFrame({'Title':Title,
                   'Procurement stage ':Procurementstage,'Noticestatus':Noticestatus,'Closing':Closing,'Contractlocation':Contractlocation,
                   'Contractvalue':Contractvalue,'Publicationdate':Publicationdate})
df.to_csv("UK.csv",index=None)


       
```
